```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
```

```python
k = int(input("enter number of folds"))
x = np.arange(1, 11).reshape(10, 1)
y = np.arange(2, 21, 2)
folds = np.array_split(np.arange(len(x)), k)
errors = []
scores=[]
for i in range(k):
  test_idx = folds[i]
  train_idx = np.concatenate([folds[j] for j in range(k) if j!= i])
  x_train = x[train_idx]
  x_test = x[test_idx]

  y_train = y[train_idx]
  y_test = y[test_idx]

  X = np.column_stack([np.ones(len(x_train)), x_train])
  beta = (X.T @ y_train) @  np.linalg.inv(X.T  @  X)

  X_test = np.column_stack([np.ones(len(x_test)), x_test])
  y_pred = X_test@beta
  mse = mean_squared_error(y_pred, y_test)
  errors.append(mse)
  a = np.sum((x_train - np.mean(x_train))**2)
  b = np.sum((y_test - y_pred)**2)
  r2_score = (a-b)/a
  scores.append(r2_score)

print(f"Total error {np.asarray(errors).mean()}")
print(f"r2_score {np.asarray(scores).mean()}")
```
